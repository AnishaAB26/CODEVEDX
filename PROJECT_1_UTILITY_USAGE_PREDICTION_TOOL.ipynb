```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load CSV
try:
    df = pd.read_csv("utility_usage.csv")
except:
    df = pd.DataFrame(columns=["Month","Usage"])
    df.to_csv("utility_usage.csv", index=False)

def train_model():
    global model
    if len(df) > 1:
        X = df[["Month"]]
        y = df["Usage"]
        model = LinearRegression()
        model.fit(X,y)

train_model()

while True:
    print("\n====== Utility Usage Prediction Tool ======")
    print("1. View Data")
    print("2. Add Data")
    print("3. Update Data")
    print("4. Predict Usage")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print(df)

    elif choice == "2":
        try:
            month = int(input("Enter Month: "))
            usage = float(input("Enter Usage: "))

            df.loc[len(df)] = [month, usage]
            df.to_csv("utility_usage.csv", index=False)

            train_model()
            print("Data Added Successfully")

        except:
            print("Invalid Input")

    elif choice == "3":
        try:
            month = int(input("Enter Month to Update: "))

            if month in df["Month"].values:
                usage = float(input("Enter New Usage: "))
                df.loc[df["Month"]==month,"Usage"]=usage
                df.to_csv("utility_usage.csv", index=False)

                train_model()
                print("Updated Successfully")

            else:
                print("Month Not Found")

        except:
            print("Invalid Input")

    elif choice == "4":
        try:
            month = int(input("Enter Month: "))

            prediction = model.predict([[month]])
            print("Predicted Usage =", round(prediction[0],2))

        except:
            print("Prediction Error")

    elif choice == "5":
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
